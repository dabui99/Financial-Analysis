```python
# library
import matplotlib.pyplot as plt
import datetime
import yfinance as yf
import pandas as pd
from scipy.signal import argrelextrema
import numpy as np
```

```python

# Defining some constants for data mining

NUM_DAYS = 10000     
INTERVAL = '1d'     
symbol = 'tsla'    

# List of symbols for technical indicators potentially used
INDICATORS = ['RSI', 'MACD', 'STOCH','ADL', 'ATR', 'MOM', 'MFI', 'ROC', 'OBV', 'CCI', 'EMV', 'BASP', 'PIVOT_FIB', 'VORTEX']


df = yf.download(symbol, period='2000d', auto_adjust = True, prepost = False, threads = True)
df.tail()
```

```python
# display Symbol price with sma21,50 and ema 20
df['sma21'] = df['Open'].rolling(window=21).mean()
df['sma50'] = df['Open'].rolling(window=50).mean()
df['ema20'] = df['Open'].ewm(span=20, adjust=False).mean()
df[['Open','sma21', 'sma50','ema20']].plot(label = symbol , figsize = (15,9))
plt.title(symbol)
```

```python
batch1 = df.loc['2022-01-01':]

n=2

ilocs_min = argrelextrema(batch1.Open.values, np.less_equal, order=n)[0]
ilocs_max = argrelextrema(batch1.Open.values, np.greater_equal, order=n)[0]

batch1.Open.plot(figsize=(15,7))
batch1.iloc[ilocs_max].Open.plot(style='.', lw=15, color='red');
batch1.iloc[ilocs_min].Open.plot(style='.', lw=15, color='green');

batch1.sma50.plot(label='sma50')
batch1.ema20.plot(label='sma100')
```
